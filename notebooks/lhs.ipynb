```python
import numpy as np

def get_random_point(start, stop, n_samples, i):
    buckets = np.linspace(start, stop, n_samples + 1)
    bucket_size = (stop - start) / n_samples
    return buckets[i] + np.random.random() * bucket_size

def get_shuffled_samples(start, stop, n_samples):
    samples = [get_random_point(start, stop, n_samples, i) for i in range(n_samples)]
    np.random.shuffle(samples)
    return samples

temperatures = get_shuffled_samples(0.1, 1.0, 10)
top_ps = get_shuffled_samples(0.0, 1.0, 10)
max_tokens = get_shuffled_samples(1000, 2000, 10)

grouped_samples = zip(temperatures, top_ps, max_tokens)

for (temp, p, max_token) in grouped_samples:
    print(temp, p, max_token)
```
